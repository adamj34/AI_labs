```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
```

```python
train_set, test_set = train_test_split(df.values, train_size=0.7, random_state=278787)
x_train = train_set[:, 0:4]
y_train = train_set[:, 4]
x_test = test_set[:, 0:4]
y_test = test_set[:, 4]
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
```

```python
plt.figure(figsize=(20, 17))
tree.plot_tree(clf);
```

```python
f"Accuracy: {clf.score(x_test, y_test)}"
```

```python
ConfusionMatrixDisplay.from_estimator(clf, x_test, y_test, cmap=plt.cm.Blues)
ConfusionMatrixDisplay.from_estimator(clf, x_test, y_test, cmap=plt.cm.Blues, normalize='true')
```
