```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
```

```python
train_set, test_set = train_test_split(df.values, test_size=0.3, random_state=278787)
x_train_set, y_train_set = train_set[:, 0:4], train_set[:, 4]
x_test_set, y_test_set = test_set[:, 0:4], test_set[:, 4]
```

```python
n_neighbors = 3
knn = KNeighborsClassifier(n_neighbors=n_neighbors, metric='euclidean')
knn.fit(x_train_set, y_train_set)
y_pred = knn.predict(x_test_set)
print(knn.score(x_test_set, y_test_set))
ConfusionMatrixDisplay.from_estimator(knn, x_test_set, y_test_set, cmap=plt.cm.Blues)
```

```python
n_neighbors = 5
knn = KNeighborsClassifier(n_neighbors=n_neighbors, metric='euclidean')
knn.fit(x_train_set, y_train_set)
y_pred = knn.predict(x_test_set)
print(knn.score(x_test_set, y_test_set))
ConfusionMatrixDisplay.from_estimator(knn, x_test_set, y_test_set, cmap=plt.cm.Blues)
```

```python
n_neighbors = 11
knn = KNeighborsClassifier(n_neighbors=n_neighbors, metric='euclidean')
knn.fit(x_train_set, y_train_set)
y_pred = knn.predict(x_test_set)
print(knn.score(x_test_set, y_test_set))
ConfusionMatrixDisplay.from_estimator(knn, x_test_set, y_test_set, cmap=plt.cm.Blues)
```

```python
gnb = GaussianNB()
gnb.fit(x_train_set, y_train_set)
y_pred = gnb.predict(x_test_set)
print(gnb.score(x_test_set, y_test_set))
ConfusionMatrixDisplay.from_estimator(gnb, x_test_set, y_test_set, cmap=plt.cm.Blues)
```
