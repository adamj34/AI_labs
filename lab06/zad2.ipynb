```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
df = pd.read_csv('iris.csv')
```

```python
train_set, test_set = train_test_split(df.values, test_size=0.3, random_state=278787)
x_train_set, y_train_set = train_set[:, 0:4], train_set[:, 4]
x_test_set, y_test_set = test_set[:, 0:4], test_set[:, 4]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# we fit the train data
scaler.fit(x_train_set)
scaler.fit(x_test_set)

x_train_set, x_test_set = scaler.transform(x_train_set), scaler.transform(x_test_set)
x_train_set
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='adam', alpha=1e-5, hidden_layer_sizes=(2,), max_iter=1500)

clf.fit(x_train_set, y_train_set)
print(clf.score(x_test_set, y_test_set))

ConfusionMatrixDisplay.from_estimator(clf, x_test_set, y_test_set)
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='adam', alpha=1e-5, hidden_layer_sizes=(3,), max_iter=1500)

clf.fit(x_train_set, y_train_set)
print(clf.score(x_test_set, y_test_set))

ConfusionMatrixDisplay.from_estimator(clf, x_test_set, y_test_set)
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='adam', alpha=1e-5, hidden_layer_sizes=(3,3), max_iter=1500)

clf.fit(x_train_set, y_train_set)
print(clf.score(x_test_set, y_test_set))

ConfusionMatrixDisplay.from_estimator(clf, x_test_set, y_test_set)
```
