```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
```

```python
nltk.download('all')
```

```python
with open('article.txt', 'r') as f:
    text = f.read()
    text = text.lower()
```

```python
# tokenize
tokens = word_tokenize(text)
print(f"Number of tokens: {len(tokens)}")
tokens
```

```python
# remove stopwords
words2del = stopwords.words('english')
words2del.extend(['ˈroːma', '[', ']'])
words2del.extend([f"[{i}]" for i in range(0, 100)])
filtered_tokens = [token for token in tokens if token not in words2del]
print(f"Number of filtered tokens: {len(filtered_tokens)}")
filtered_tokens
```

```python
# lematize
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
print(f"Number of lemmatized tokens: {len(lemmatized_tokens)}")
lemmatized_tokens
```

```python
# Join the tokens back into a string
processed_text = ' '.join(lemmatized_tokens)
processed_text
```

```python
unique_words = set(lemmatized_tokens)
word_count = sorted([(word, lemmatized_tokens.count(word)) for word in unique_words], key=lambda x: x[1], reverse=True)
word_count
```

```python
top10words = [el[0] for el in word_count[:10]]
top10lenghts = [el[1] for el in word_count[:10]]
plt.bar(top10words, top10lenghts);
```

```python
wordcloud = WordCloud().generate(processed_text)
plt.imshow(wordcloud, interpolation='bilinear');
```
